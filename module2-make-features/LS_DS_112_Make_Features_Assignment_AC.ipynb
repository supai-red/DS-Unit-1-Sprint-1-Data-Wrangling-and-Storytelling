```python
def term_to_int(cell):
  return int(cell.strip(' months'))
```

```python
df['term_int'] = df['term'].apply(term_to_int)

df.head()
```

Make a column named loan_status_is_great. 

It should contain the integer 1 if loan_status is "Current" or "Fully Paid."
 
Else it should contain the integer 0.

```python
def loan_stat(cell):
  """Function to determine if loan is current or fully paid."""
  if cell == "Current" or cell == "Fully Paid":
    return 1
  else:
    return 0  
```

```python
df['loan_status_is_great'] = df['loan_status'].apply(loan_stat)
```
